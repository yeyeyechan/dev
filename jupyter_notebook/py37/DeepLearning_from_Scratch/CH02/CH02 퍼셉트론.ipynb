## <span style="color:DarkCyan">2.2 단순한 논리 회로 </span>


### <span style="color:indigo">2.2.1 AND 게이트 </span><br><br>


AND 게이트는 입력이 둘이고 출력은 하나다 <br><br>


```python
img = imread('AND.png')
plt.imshow(img)
plt.show()
```

위와 같은 입력신호와 출력신호의 대응 표를 <span style="color:red">진리표</span> 라고 한다. <br><br>

이 AND 게이트를 퍼셉트론으로 표현하고 싶다면 <span style="color:red">진리표</span> 대로 작동하는 w,theta의 값을 설정하면 된다.<br><br>
위의 진리표를 만족하는 매개변수 조합은 무한이 많다. <br><br>
예를 들면 (0.5 , 0.5, 0.8) (1.0, 1.0 , 1.0) 등이 있다.<br><br>

### <span style="color:indigo">2.2.2 NAND 게이트 와 OR 게이트</span><br><br>


NAND 게이트란 Not AND 를 의미하며 그 동작은 AND 게이트의 출력을 뒤집은 것이 된다. <br><br>
아래는 NAND 게이트의 진리표이다 <br><br>

```python
img = imread('NAND.png')
plt.imshow(img)
plt.show()
```

이를 표현 할 수있는 매개변수 조합은 무한이 많다 예를들어 <br><br>
(-0.5, -0.5, -0.7) <br><br>

OR 게이트는 입력 신호 중 하나 이상이 1이면 출력이 1이 되는 논리 회로이다.

```python
img = imread('OR.png')
plt.imshow(img)
plt.show()
```

<span style="color:red">기계학습</span> 문제는 이 <span style="color:red">매개변수</span> 값을 정하는 작업을 컴퓨터가 자동으로 하도록 한다. <br><br>
<span style="color:red">학습</span>이란 적절한 매개변수 값을 정하는 작업이며<br><br>
사람은 퍼셉트론의 구조를 고민하고 컴퓨터에 학습할 데이터를 주는 일을 한다.<br><br>

이상과 같이 퍼셉트론으로 위의 논리 회로를 모두 표현 할수 있다. <br><br>
여기서 중요한점은 퍼셉트론의 구조는 위 3가지 게이트에서 모두 같지만 <span style="color:red">매개변수(가중치와 임계값)</span>이 다르다는 것이다.<br><br>

## <span style="color:DarkCyan">2.3 퍼셉트론 구현하기 </span>


### <span style="color:indigo">2.3.1 간단한 구현 </span><br><br>


```python
def AND(x1, x2):
    w1,w2, theta = 0.5, 0.5,0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```
